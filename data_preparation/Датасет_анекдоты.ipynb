```python
from natasha import (
    Segmenter,
    MorphVocab,
    
    NewsEmbedding,
    NewsMorphTagger,
    NewsSyntaxParser,
    NewsNERTagger,
    
    PER,
    NamesExtractor,

    Doc
)
```

```python
segmenter = Segmenter()
morph_vocab = MorphVocab()

emb = NewsEmbedding()
morph_tagger = NewsMorphTagger(emb)
syntax_parser = NewsSyntaxParser(emb)
ner_tagger = NewsNERTagger(emb)

names_extractor = NamesExtractor(morph_vocab)
```

```python
# Create а text file

with open("all_data_with_keywords.txt", "w") as file:
    file.write("")
```

```python
def add_key_words(dataset):
    with open("all_data_with_keywords.txt", "a", encoding="UTF-8") as file:
        keys_for_dataset = []
        for i, text in enumerate(dataset):
            key_words = []
            doc = Doc(text)
            doc.segment(segmenter)
            doc.tag_morph(morph_tagger)
            doc.parse_syntax(syntax_parser)
            for token in doc.tokens:
                if token is not None:
                    token.lemmatize(morph_vocab)
            key_words.extend([i.lemma for i in doc.tokens if (i.rel=='nsubj' or i.rel=='obl' or i.rel=='obj' or i.rel=='conj') and i.pos!='PRON'])
            keys_for_dataset.append(key_words)
            file.writelines( f"|0|1|Расскажи анекдот про {' '.join(list(set(key_words)))}|1|-|\"{text}\"|\n")
```

```python
keywords = []
for i, item in enumerate(dataset):
  new = item.split('|1|-|\"')
  if len(new) > 1:
    dataset[i] = new[1]
    keywords.append(new[0])
  else:
    dataset[i] = new
    keywords.append([])
dataset
```

```python
all = list(zip(keywords, dataset))
all
```
